### **1: Social Media – Post Engagement Analyzer**

```python
def post_engagement(likes):
    total_likes = sum(likes)

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)
post_engagement([200, 300, 250, 300])
```

### **3: Agriculture – Rainfall Adequacy Checker**


```python
def rainfall_checker(rainfall, required):
    avg = sum(rainfall) / len(rainfall)

    if avg >= required:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", int(avg))
    print("Rainfall Status:", status)

rainfall_checker([70, 80, 65, 75], 70)
```

### **4: Social Media – Duplicate Account Detection**

```python
def duplicate_accounts(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

duplicate_accounts(["user1", "user2", "user1"])
```

### **6: Agriculture – Premium Crop Price Filter**

```python
def premium_crops(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)

premium_crops([1500, 2500, 3200, 1800])
```

### **8: Banking – Daily Transaction Limit Checker**

```python
def transaction_check(amount):
    limit = 50000

    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

transaction_check(60000)
```

### **9: E-Learning – Student Attendance Eligibility System**

```python
def attendance_check(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance_check(["P", "P", "A", "P", "P"])
```

### **10: Smart Electricity Bill Analyzer**

```python
def electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3

    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5

    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Bill:", bill)
    print("Usage Level:", usage)

electricity_bill(250)
```
