### **1: Unique Words in a Sentence**

```python
#Taking input as
sentence = input("Enter sentence: ")
words = sentence.split()
unique_words = set(words) #removing duplicates
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

### **2: Highest Salary from Employee Data**

```python
#Taking input as
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = ""
highest_salary = 0

for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_employee = name

print("Highest Salary:", highest_employee,"=",highest_salary)
```

### **3: Find Maximum and Minimum Values**

```python
#Taking input as
numbers = [45, 22, 89, 10, 66]
max_val = numbers[0]
min_val = numbers[0]
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

### **4: Count Products Above a Price Threshold**

```python
#Taking input as
prices = [450, 1200, 899, 1500, 300]
count = 0
for price in prices:
    if price > 1000:
        count = count + 1
print("Products above 1000:", count)
```

### **5: Calculate Attendance Percentage**

```python
#Taking input as
attendance = ["P", "P", "A", "P", "P"]
present = 0
for day in attendance:
    if day == "P":
        present = present + 1
percentage = (present / len(attendance)) * 100
print("Attendance Percentage:", percentage)
```

### **6: Remove Duplicate Phone Numbers**

```python
#Taking input as
phone_numbers = [9876543210, 9123456789, 9876543210]
unique_numbers = set(phone_numbers)
print("Unique phone numbers:", unique_numbers)
```

### **7: Count Character Frequency**

```python
#Taking input as
text = "python"
freq = {}
for ch in text:
    if ch in freq:
        freq[ch] = freq[ch] + 1
    else:
        freq[ch] = 1
print("Characters are :",freq)
```

### **9: Check if a Key Exists in a Dictionary**

### **10: Calculate Average Marks**

```python
#Taking input as
marks = [70, 80, 65, 75]
total = 0

for m in marks:
    total = total + m

average = total / len(marks)
print("Average Marks:", average)
```
