### **1: Smart Light Controller**

```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"
    def turn_on(self):
        self.status = "ON"
        print(f"{self.light_name} is {self.status}")
    def turn_off(self):
        self.status = "OFF"
        print(f"{self.light_name} is {self.status}")
light = SmartLight("Bedroom Light")
light.turn_on()
```

### **Employee ID Card System**

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)
emp = Employee("Rahul", "EMP102", "IT") #input taken as
emp.display_id_card()
```

### **3: Mobile Contact Record**

```python
class contactlist:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    def save_contactlist(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)
contact = contactlist("Anita", 9876543210)
contact.save_contactlist()
```

### **4: Product Price Tag Generator**

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def price_tag(self):
        print("Product:", self.name)
        print(f"Price: ₹{self.price}")
product = Product("Headphones", 2499)
product.price_tag()
```

### **5: Movie Rating Display System**

```python
class Movie:
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating
    def display_movie(self):
        print("Movie:", self.movie_name)
        print(f"Rating: {self.rating} / 5")
movie = Movie("Inception", 4.8)
movie.display_movie()
```

### **Delivery Address Manager**

```python
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address
    def delivery_details(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)
order = Delivery("Suman", "Hyderabad") #input taken as
order.delivery_details()
```
