**2. Pass / Fail Analyzer**

```python

#Given:
marks = [45, 78, 90, 33, 60]

#fixing value
Student_passed = 0
Student_failed = 0

for mark in marks:
    if mark >= 50:
        Student_passed = Student_passed + 1
    else:
        Student_failed = Student_failed + 1

print("Total Passed Students:", Student_passed)
print("Total Failed Students:", Student_failed)

#output:
```

**3. Simple Data Cleaner**

```python
#Given:
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for naming in names:
    clean_name = naming.strip().lower()
    cleaned_names.append(clean_name)

print("Cleaned List:", cleaned_names)

#output:
```

**4. Message Length Analyzer**

```python
#Given:
messages = ["Hi", "Welcome to the platform", "OK"]

for text in messages:
    length = len(text)
    print("Message:", text)
    print("Length:", length)

    if length > 10:
        print("This message is longer than 10 characters")

    print()

#output:
```

**5. Error Message Detector**

```python
#Given:
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for error in logs:
    if error == "ERROR":
        error_count = error_count + 1

print("Total ERROR entries:", error_count)

#output:
```
