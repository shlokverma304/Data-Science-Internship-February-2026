### **1: Mobile Recharge Validation System**

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: "))

        if amount < 50:
            print("Recharge amount must be at least 50")
        elif amount not in valid_plans:
            print("Invalid plan! Valid plans are:", valid_plans)
        else:
            print("Recharge Successful!")
            break

validate_recharge()
```

### **2: Inventory Reorder Alert System**

```python
def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")

products = {
    "Rice": 50, #(>15)
    "Sugar": 25, #(>15)
    "Oil": 8, #(<15)
    "Wheat": 5 #(<15)
}

check_inventory(products)
```

### **3: Student Result Processing System**

```python
def student_score(marks):
    total = 0
    for m in marks:
        total += m

    average = total / len(marks)
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

marks_list = [60, 45, 70, 50]
print("Result:", student_score(marks_list))
```

### **4: Cab Fare Estimator with Retry Option**

```python
def calculate_charge(distance, peak):
    charge = 50 + (12 * distance)

    if peak == "yes":
        charge += charge * 0.25
    return charge

while True:
    km = float(input("Enter distance (km): "))
    peak_hour = input("Is it peak hour? (yes/no): ")

    total_charge = calculate_charge(km, peak_hour)
    print("Total Fare: ", total_charge)

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

### **5: Employee Attendance Eligibility**

```python
def check_eligibility(attendance):
    present = 0
    for day in attendance:
        if day == "P":
            present += 1
    percentage = (present / len(attendance)) * 100
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
attendance_list = ["P", "P", "A", "P", "A", "P", "P", "P"]
print("Attendance Status:", check_eligibility(attendance_list))
```

### **6: Password Strength Checker**

```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak Password"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

pwd = input("Enter password: ")
print(check_password(pwd))
```
